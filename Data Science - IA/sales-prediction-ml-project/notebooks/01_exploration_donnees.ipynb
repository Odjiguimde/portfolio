## 4️⃣ Détection des anomalies

```python
# Valeurs manquantes
print("⚠️  VALEURS MANQUANTES")
print("="*50)
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Nombre': missing,
    'Pourcentage': missing_pct
})
print(missing_df[missing_df['Nombre'] > 0])
print("="*50)
```

```python
# Prix aberrants
print("💸 DÉTECTION DES PRIX ABERRANTS")
print("="*50)

# Prix très bas (< 100 FCFA)
prix_bas = df[df['prix'] < 100]
print(f"Prix < 100 FCFA : {len(prix_bas)} lignes ({len(prix_bas)/len(df)*100:.2f}%)")
if len(prix_bas) > 0:
    print("\nExemples :")
    print(prix_bas[['date', 'produit', 'prix', 'quantite']].head())

# Prix très élevés (> 1 000 000 FCFA)
prix_eleves = df[df['prix'] > 1000000]
print(f"\nPrix > 1 000 000 FCFA : {len(prix_eleves)} lignes ({len(prix_eleves)/len(df)*100:.2f}%)")
if len(prix_eleves) > 0:
    print("\nExemples :")
    print(prix_eleves[['date', 'produit', 'prix', 'quantite']].head())

print("="*50)
```

```python
# Quantités aberrantes
print("📊 DÉTECTION DES QUANTITÉS ABERRANTES")
print("="*50)

# Quantités négatives ou nulles
qty_negative = df[df['quantite'] <= 0]
print(f"Quantités <= 0 : {len(qty_negative)} lignes ({len(qty_negative)/len(df)*100:.2f}%)")
if len(qty_negative) > 0:
    print("\nExemples :")
    print(qty_negative[['date', 'produit', 'prix', 'quantite']].head())

# Quantités très élevées (> 1000)
qty_elevees = df[df['quantite'] > 1000]
print(f"\nQuantités > 1000 : {len(qty_elevees)} lignes ({len(qty_elevees)/len(df)*100:.2f}%)")
if len(qty_elevees) > 0:
    print("\nExemples :")
    print(qty_elevees[['date', 'produit', 'categorie', 'quantite']].head())

print("="*50)
```

```python
# Doublons
print("🔄 DÉTECTION DES DOUBLONS")
print("="*50)

doublons = df[df.duplicated(keep=False)]
print(f"Nombre de lignes dupliquées : {len(doublons)} ({len(doublons)/len(df)*100:.2f}%)")
print(f"Nombre de groupes de doublons : {len(doublons)//2}")

if len(doublons) > 0:
    print("\nExemples de doublons :")
    print(doublons.sort_values('date').head(6))

print("="*50)
```

## 5️⃣ Analyses Business

```python
# Top 10 produits par chiffre d'affaires
print("🏆 TOP 10 PRODUITS PAR CHIFFRE D'AFFAIRES")
print("="*70)

top_ca = df.groupby('produit')['chiffre_affaires'].sum().sort_values(ascending=False).head(10)

for i, (produit, ca) in enumerate(top_ca.items(), 1):
    pct = (ca / df['chiffre_affaires'].sum()) * 100
    print(f"{i:2d}. {produit:40s} : {ca:15,.0f} FCFA ({pct:5.2f}%)")

print("="*70)
```

```python
# CA par ville
print("🌍 CHIFFRE D'AFFAIRES PAR VILLE")
print("="*70)

ca_ville = df.groupby('ville').agg({
    'chiffre_affaires': ['sum', 'mean', 'count']
}).round(0)

ca_ville.columns = ['CA Total', 'CA Moyen', 'Nb Transactions']
ca_ville = ca_ville.sort_values('CA Total', ascending=False)
ca_ville['% du CA Total'] = (ca_ville['CA Total'] / ca_ville['CA Total'].sum() * 100).round(2)

print(ca_ville)
print("="*70)
```

```python
# CA par catégorie
print("📦 CHIFFRE D'AFFAIRES PAR CATÉGORIE")
print("="*70)

ca_categorie = df.groupby('categorie').agg({
    'chiffre_affaires': ['sum', 'mean', 'count']
}).round(0)

ca_categorie.columns = ['CA Total', 'CA Moyen', 'Nb Transactions']
ca_categorie = ca_categorie.sort_values('CA Total', ascending=False)
ca_categorie['% du CA Total'] = (ca_categorie['CA Total'] / ca_categorie['CA Total'].sum() * 100).round(2)

print(ca_categorie)
print("="*70)
```

```python
# CA par mode de paiement
print("💳 CHIFFRE D'AFFAIRES PAR MODE DE PAIEMENT")
print("="*70)

ca_paiement = df.groupby('mode_paiement').agg({
    'chiffre_affaires': ['sum', 'mean', 'count']
}).round(0)

ca_paiement.columns = ['CA Total', 'CA Moyen', 'Nb Transactions']
ca_paiement = ca_paiement.sort_values('CA Total', ascending=False)
ca_paiement['% du CA Total'] = (ca_paiement['CA Total'] / ca_paiement['CA Total'].sum() * 100).round(2)

print(ca_paiement)
print("="*70)
```

```python
# Évolution mensuelle du CA
print("📅 ÉVOLUTION MENSUELLE DU CHIFFRE D'AFFAIRES")
print("="*70)

# Extraction du mois
df['mois'] = df['date'].dt.to_period('M')

# CA mensuel
ca_mensuel = df.groupby('mois').agg({
    'chiffre_affaires': ['sum', 'mean', 'count']
}).round(0)

ca_mensuel.columns = ['CA Total', 'CA Moyen', 'Nb Transactions']

print(ca_mensuel)
print("="*70)
```

## 7️⃣ Analyse de la saisonnalité

Le marché sénégalais est fortement impacté par les fêtes religieuses et événements culturels.

```python
# Identification des périodes spéciales
periodes_speciales = {
    'Ramadan': ('2024-03-11', '2024-04-09'),
    'Korité': ('2024-04-10', '2024-04-15'),
    'Tabaski': ('2024-06-10', '2024-06-20'),
    'Magal Touba': ('2024-08-20', '2024-08-25'),
    'Gamou': ('2024-09-15', '2024-09-20'),
    'Rentrée': ('2024-10-01', '2024-10-15'),
    'Fin année': ('2024-12-15', '2024-12-31')
}

print("🎉 ANALYSE DES PÉRIODES SPÉCIALES")
print("="*70)

ca_normal_list = []
ca_periode_list = []

for periode, (debut, fin) in periodes_speciales.items():
    mask = (df['date'] >= debut) & (df['date'] <= fin)
    ca_periode = df[mask]['chiffre_affaires'].sum()
    nb_jours = (pd.to_datetime(fin) - pd.to_datetime(debut)).days + 1
    ca_jour_periode = ca_periode / nb_jours
    
    # CA moyen journalier hors période
    mask_hors = ~mask
    ca_hors = df[mask_hors]['chiffre_affaires'].sum()
    nb_jours_hors = len(df[mask_hors]['date'].dt.date.unique())
    ca_jour_normal = ca_hors / nb_jours_hors if nb_jours_hors > 0 else 0
    
    boost = ((ca_jour_periode / ca_jour_normal) - 1) * 100 if ca_jour_normal > 0 else 0
    
    print(f"\n{periode} ({debut} → {fin}) :")
    print(f"  CA total période : {ca_periode:,.0f} FCFA")
    print(f"  CA/jour période : {ca_jour_periode:,.0f} FCFA")
    print(f"  CA/jour normal : {ca_jour_normal:,.0f} FCFA")
    print(f"  📈 Boost : {boost:+.1f}%")

print("\n" + "="*70)
```
